# 매개변수 곡선(Parametric Curve)

$N$차원 공간에서 어떤 그래프 $c(t):\mathbb{R}\mapsto\mathbb{R^n}$는 다음과 같이 정의된다:
$$ c(t)= \begin{bmatrix}f_1(t)\\f_2(t)\\ \vdots \\ f_n(t)\end{bmatrix}$$
여기서 $c(t)$를 $t$에 대한 매개변수 곡선이라 한다.

2차원 공간의 매개변수 곡선은 두 개의 함수로 정의할 수 있다. 

$$c(t) =\begin{bmatrix}f(t)\\g(t)\end{bmatrix}$$

3차원 공간의 매개변수 곡선은 세 개의 함수로 정의할 수 있다.

$$c(t) = \begin{bmatrix}f(t)\\ g(t)\\ h(t) \end{bmatrix}$$


## 원의 매개변수 곡선

$\theta\in\mathbb{R}$에 대하여 반지름이 $r$인 원의 매개변수 곡선 $c(\theta):\mathbb{R}\mapsto\mathbb{R}^2$는 다음과 같이 정의되며, 주기가 $2\pi$인 주기함수이다.

$$
c(\theta) = \begin{bmatrix}r\cos{\theta}\\r\sin{\theta}\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot()

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
r = 5
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.annotate("(3,4)", [3, 4], [4,5],arrowprops=dict(facecolor='black', shrink=0.01))
ax.plot(x, y)
```

## 가로로 누운 이차함수의 매개변수 곡선

어떤 2차함수의 그래프 $(x, y)$에서 $y=2x-2x^2$일 때, 이를 매개변수 방정식으로 나타내면 다음과 같다.

$$
c(t) = 
\begin{bmatrix}
t\\
2t-2t^2 
\end{bmatrix}
$$

여기서 양변에 90도 회전 행렬을 곱하면 원점을 기준으로 반시계방향으로 90도 회전한 그래프를 얻을 수 있다.

$$
R_{90\degree}c(t) = 
R_{90\degree}
\begin{bmatrix}
t\\
2t-2t^2 
\end{bmatrix}
=
\begin{bmatrix}
0 & -1\\
1 & 0
\end{bmatrix}
\begin{bmatrix}
t\\
2t-2t^2 
\end{bmatrix}
=
\begin{bmatrix}
2t^2 - 2t\\
t
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot()

t = np.linspace(-1, 2, 50)
x = t
y = 2*t - 2*t**2

ax.plot(x, y, label='$c(t)$')

t = np.linspace(-1, 2, 50)
x = 2*t**2 - 2*t
y = t
ax.plot(x, y, label='$R_{90\degree}c(t)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.set_aspect('equal')
ax.legend()
```
